### 毎日が天皇誕生日になるまでかかる日数のシミュレーション（国民の祝日考慮ver.）

```python
%pylab inline
```

```python
import random
import collections
import numpy as np
import matplotlib.pyplot as plt

result = [[] for t in range(1200)]

for z in range(1000):
    N = 365
    count = 0
    days = [0] * N
    for t in range(1200):
        i = random.randint(0, N - 1)
        if days[i] == 0:
            days[i] = 1
            count += 1
            if days[i - 2] == 1 and days[i - 1] == 0:
                days[i - 1] = 1
                count += 1
            if days[(i + 2) % N] == 1 and days[(i + 1) % N] == 0:
                days[(i + 1) % N] = 1
                count += 1
        result[t].append(count)

data = np.array(result)
mean = data.mean(axis=1)
quantile1 = np.percentile(data, 25, axis=1)
quantile3 = np.percentile(data, 75, axis=1)
plt.xlabel('#change')
plt.ylabel('#holiday')
plt.plot(range(1, 1201), mean, lw=2)
plt.fill_between(range(1, 1201), quantile1, quantile3, facecolor='yellow', alpha=0.5)
plt.grid()
plt.show()
```
