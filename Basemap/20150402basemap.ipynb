```python
from mpl_toolkits.basemap import Basemap
```

## World Map

```python
fig = plt.figure(figsize=(15,15))
mymap = Basemap(lat_0=35, lon_0=135)
mymap.drawcoastlines()
mymap.drawcountries()
mymap.fillcontinents(color='coral', alpha=0.3)
mymap.drawmapboundary()
mymap.drawmeridians(np.arange(0,360,15))
mymap.drawparallels(np.arange(-90,90,15))
plt.show()
```

## Japanese Map

```python
fig = plt.figure(figsize=(8,8))
mymap2 = Basemap(projection='merc',
                 resolution='h',
                 llcrnrlon=110,
                 llcrnrlat=20,
                 urcrnrlon=160,
                 urcrnrlat=60)
mymap2.drawcoastlines()
mymap2.drawcountries()
mymap2.fillcontinents(color='coral', alpha=0.3)
mymap2.drawmapboundary()
plt.show()
```

## Awaji-Island & Hiratsuka-city

```python
fig = plt.figure(figsize=(8,8))
m = Basemap(projection='merc',
                 resolution='h',
                 llcrnrlon=132,
                 llcrnrlat=33,
                 urcrnrlon=140,
                 urcrnrlat=38)
m.drawcoastlines(linewidth=0.5, color='k')
m.fillcontinents(color='#eeeeee',lake_color='#ddeeff')
m.drawstates( linewidth=0.5, color='k' )
m.drawcountries()
m.drawmapboundary(fill_color='#ddeeff')
m.drawparallels(np.arange(31.0, 38.1, 1.0), labels = [1,0,0,0], fontsize=12)
m.drawmeridians(np.arange(132.0, 140.1, 2.0), labels = [0,0,0,1], fontsize=12)

sumoto_lon = 134.7687
sumoto_lat = 34.3755
hiratsuka_lon = 139.3581
hiratsuka_lat = 35.3587

x1,y1 = m(sumoto_lon, sumoto_lat)
x2,y2 = m(hiratsuka_lon, hiratsuka_lat)
m.plot(x1, y1, 'bo', markersize=10)
m.plot(x2, y2, 'bo', markersize=10)
plt.text(x1 + 10000, y1 + 10000, u"sumoto")
plt.text(x2 + 10000, y2 + 10000, u"hiratsuka")

plt.show()
```
