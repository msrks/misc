```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import rfft, rfftfreq
```

```python
time = np.arange(0,1,0.005)
```

## 1) freq = 50Hzの信号

```python
f1 = 50
signal_1 = np.sin(2*np.pi*f1*time)
plt.plot(time, signal_1)
plt.xlabel('time')
plt.ylabel('intensity')
```

```python
freq = rfftfreq(len(signal_1), 0.005)
Sig_f1 = rfft(signal_1)
Int_f1 = np.sqrt(np.real(Sig_f1)**2+np.imag(Sig_f1)**2)
plt.plot(freq, Int_f1)
plt.xlabel('freq(Hz)')
plt.ylabel('intensity')
```

## 2) freq = 52Hzの信号

```python
f2 = 52
signal_2 = np.sin(2*np.pi*f2*time)
plt.plot(time, signal_2)
plt.xlabel('time')
plt.ylabel('intensity')
```

```python
Sig_f2 = rfft(signal_2)
Int_f2 = np.sqrt(np.real(Sig_f2)**2+np.imag(Sig_f2)**2)
plt.plot(freq, Int_f2)
plt.xlabel('freq(Hz)')
plt.ylabel('intensity')
```

## 3) freq = 50Hzの信号と freq = 52Hz の 信号をかけたら、2Hzの信号が取り出せる（down conversion)

```python
signal_3 = signal_1*signal_2
plt.plot(time, signal_3)
plt.xlabel('time')
plt.ylabel('intensity')
```

```python
Sig_f3 = rfft(signal_3)
Int_f3 = np.sqrt(np.real(Sig_f3)**2+np.imag(Sig_f3)**2)
plt.plot(freq, Int_f3)
plt.xlabel('freq(Hz)')
plt.ylabel('intensity')
```
